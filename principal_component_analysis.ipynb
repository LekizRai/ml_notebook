```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA as pca
```

```python
X, _ = make_blobs(n_samples=300, centers=1, cluster_std=0.6)
X[:, 1] /= 3
```

```python
def rotate(X, deg=90):
    rad = (deg / 180) * np.pi
    rotation_matrix = np.array([[np.cos(rad), np.sin(rad)], [-np.sin(rad), np.cos(rad)]])
    return X.dot(rotation_matrix)

def scatter_hist(X, ax, ax_histx, ax_histy):
    ax_histx.tick_params(axis="x", labelbottom=False)
    ax_histy.tick_params(axis="y", labelleft=False)

    ax.scatter(X[:, 0], X[:, 1])
    
    ax_histx.hist(X[:, 0], bins=30, color='black')
    ax_histy.hist(X[:, 1], bins=30, color='black', orientation='horizontal')
    
fig = plt.figure(figsize=(6, 6))
gs = fig.add_gridspec(2, 2, width_ratios=(4, 1), height_ratios=(1, 4), 
                      left=0.1, right=0.9, bottom=0.1, top=0.9,
                      wspace=0.05, hspace=0.05)

X = rotate(X, 45)
x_mean, y_mean = np.mean(X, axis=0)

ax = fig.add_subplot(gs[1, 0])
ax.set_xlim(x_mean - 2, x_mean + 2)
ax.set_ylim(y_mean - 2, y_mean + 2)
ax.set_aspect('equal')

ax_histx = fig.add_subplot(gs[0, 0], sharex=ax)
ax_histy = fig.add_subplot(gs[1, 1], sharey=ax)

scatter_hist(X, ax, ax_histx, ax_histy)
```

```python
def PCA(X, dimension=10):
    N = X.shape[0]
    X_std = X - np.mean(X, axis=0, keepdims=True)
    S = (1 / N) * ((X.T).dot(X))
    eigen_values, eigen_vectors = np.linalg.eig(S)
    sorted_id = np.argsort(-1 * eigen_values)
    id = sorted_id[:dimension]
    U_k = eigen_vectors[:, id].reshape(X.shape[1], dimension)
    return (U_k.T).dot(X_std.T)  
```

```python
model = pca(1)
model.fit(X)
y = model.transform(X).reshape(1, 300)

# Difference between PCA by Scikit-learn and PCA self-made
print(np.sum((np.sort(PCA(X, 1)) - np.sort(y)) ** 2) / X.shape[0])
```
