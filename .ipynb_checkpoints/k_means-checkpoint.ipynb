```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.6, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=20, c=y)
plt.title("Initial data")
```

```python
def get_euclid_distance(x, y):
  return np.sum((x - y) ** 2)

def assign_label(x, center_points):
  return np.argmin([get_euclid_distance(x, point) 
                    for point in center_points])

def find_center(X, y, center_points):
  for i in range(len(center_points)):
    if np.sum(y == i) == 0:
      continue
    center_points[i, :] = np.mean(X[y == i, :], axis=0)

def check_stop(new_centers, old_centers):
  return np.sum(np.array([np.sum(abs(new_centers[i] - old_centers[i])) 
                for i in range(len(new_centers))])) < 0.001
    
```

```python
def k_means_fit(X, K):
  center_points = X[np.random.randint(X.shape[0], size=K)]
  new_centers = X[np.random.randint(X.shape[0], size=K)]
  y = None
  iteration = 0
  while not check_stop(new_centers, center_points):
    y = np.array([assign_label(x, center_points) for x in X])
    new_centers = center_points.copy()
    find_center(X, y, center_points)
    iteration += 1
  print("Number of iterations: ", iteration, '\n')
  return y, center_points
```

```python
y, center_points = k_means_fit(X, 4)
print(center_points)
plt.scatter(X[:, 0], X[:, 1], c=y, s=20)
plt.scatter(center_points[:, 0], center_points[:, 1], c='r')
```
