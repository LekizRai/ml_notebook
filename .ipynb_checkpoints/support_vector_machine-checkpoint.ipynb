```python
import numpy as np
import matplotlib.pyplot as plt
from cvxopt import matrix, solvers

# So cool settings
square_setting = {"marker": "s", "facecolor": "none", "edgecolor": "black"}
circle_setting = {"marker": "o", "color": "black"}
main_line = {"linestyle": "-", "color": "black"}
supporting_line = {"linestyle": "--", "linewidth": 0.5, "color": "black"}
```

```python
np.random.seed(22)

means = [[2, 2], [4, 2]]
cov = [[.3, .2], [.2, .3]]
N = 10
X0 = np.random.multivariate_normal(means[0], cov, N)
X1 = np.random.multivariate_normal(means[1], cov, N)
X = np.concatenate([X0, X1], axis=0)
y = np.concatenate([np.ones(N), -np.ones(N)], axis=0)
```

```python
# How to determine sign of y_n ??

def SVM(X, y):
    unique_y = np.unique(y)
    id_0 = np.where(y == unique_y[0])
    id_1 = np.where(y == unique_y[1])
    y[id_0] = -1
    y[id_1] = 1
    n = X.shape[0]
    V = np.concatenate([-X[id_0], X[id_1]], axis=0)
    Q = matrix(V.dot(V.T))
    p = matrix(-np.ones((n, 1)))
    G = matrix(-np.eye(n))
    h = matrix(np.zeros((n, 1)))
    A = matrix(y.reshape(1, -1))
    b = matrix(np.zeros((1, 1)))
    solvers.options['show_progress'] = False
    sol = solvers.qp(Q, p, G, h, A, b)
    l = np.array(sol['x'])

    w_dual = V.T.dot(l)
    S = np.where(l > 1e-8)[0]
    b_dual = np.mean(y[S].reshape(-1, 1) - X[S, :].dot(w_dual))
    return w_dual, b_dual
```

```python
w, b = SVM(X, y)
print(w.T)
print(b)
```

```python
plt.scatter(X0[:, 0], X0[:, 1], **square_setting)
plt.scatter(X1[:, 0], X1[:, 1], **circle_setting)
x_plot = np.linspace(1, 5, 50)
y_plot = (w[0, 0] / -w[1, 0]) * x_plot + (b / -w[1, 0])

y_first_margin = (w[0, 0] / -w[1, 0]) * x_plot + ((b + 1) / -w[1, 0])
y_second_margin = (w[0, 0] / -w[1, 0]) * x_plot + ((b - 1) / -w[1, 0])

plt.plot(x_plot, y_plot, **main_line)
plt.plot(x_plot, y_first_margin, **supporting_line)
plt.plot(x_plot, y_second_margin, **supporting_line)
plt.fill_between(x_plot, 10, y_plot, facecolor='none', hatch='x', alpha=0.1)
plt.fill_between(x_plot, y_plot, facecolor='none', hatch='+', alpha=0.1)
plt.xlim(1, 5)
plt.ylim(0, 3)
```
