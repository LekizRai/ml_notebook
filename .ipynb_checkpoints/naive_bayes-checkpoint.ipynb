```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.randn(1000, 2) * 10
y = np.zeros(1000)
y[X[:, 1] > 0.5 * X[:, 0]] = 1
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Initial data")
```

```python
def probability_calc(xi, sigmaci_2, muci):
  left_part = 1 / np.sqrt(2 * np.pi * sigmaci_2)
  right_part = np.exp(((xi - muci) ** 2) / (-2 * sigmaci_2))
  return left_part * right_part
```

```python
def NB_fit(X, y):
    unique_y = np.unique(y)
    mu = np.array([[np.mean(X[y == unique_y[i], j]) for j in range(X.shape[1])]
               for i in range(len(unique_y))])
    sigma_2 = np.array([[np.std(X[y == unique_y[i], j]) for j in range(X.shape[1])]
                  for i in range(len(unique_y))]) ** 2
    p_c = np.array([(len(X[y == unique_y[i]]) / len(X)) 
                for i in range(len(unique_y))])
    return mu, sigma_2, p_c
```

```python
X_test = np.random.rand(200, 2) * 10 - 4
y_test = np.zeros(200)
mu, sigma_2, p_c = NB_fit(X, y)
for it in range(200):
  x = X_test[it]
  id = np.argmax([np.log(p_c[i]) 
                        + np.sum([np.log(probability_calc(x[j], sigma_2[i, j],   
                                                          mu[i, j])) 
                                  for j in range(X.shape[1])])
                  for i in range(len(p_c))])
  y_test[it] = id
  
result = np.array([1 if (X_test[i, 1] >= 0.5 * X_test[i, 0] and y_test[i] == 1)
                        or (X_test[i, 1] <= 0.5 * X_test[i, 0] and y_test[i] == 0) 
                     else 0 for i in range(len(X_test))])
```

```python
print(X_test[0])
print(y_test[0])
print(np.mean(result))
plt.scatter(X[:, 0], X[:, 1], c=y, s=5)
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, s=5)
plt.title("Result after testing")
```
