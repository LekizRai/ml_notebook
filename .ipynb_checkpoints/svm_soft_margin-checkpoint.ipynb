```python
import numpy as np
import matplotlib.pyplot as plt

from cvxopt import matrix, solvers

# So cool settings
square_setting = {"marker": "s", "facecolor": "none", "edgecolor": "black"}
circle_setting = {"marker": "o", "color": "black"}
main_line = {"linestyle": "-", "color": "black"}
supporting_line = {"linestyle": "--", "linewidth": 0.5, "color": "black"}
```

```python
np.random.seed(22)

means = [[2, 2], [4, 2]]
cov = [[.7, 0], [0, .7]]
N = 20
X0 = np.random.multivariate_normal(means[0], cov, N)
X1 = np.random.multivariate_normal(means[1], cov, N)
X = np.concatenate([X0, X1], axis=0)
y = np.concatenate([np.ones(N), -np.ones(N)], axis=0)
```

```python
def SVM_soft_margin(X, y, C=100):
    # Copy matrix y
    y_svm = y.copy()
    
    # Transform y into array of 1 and -1
    unique_y = np.unique(y)
    id_0 = np.where(y == unique_y[0])
    id_1 = np.where(y == unique_y[1])
    y_svm[id_0] = -1
    y_svm[id_1] = 1
    
    # Convex optimization problem solving
    n = X.shape[0]
    V = X.copy()
    V[id_0] *= -1
    Q = matrix(V.dot(V.T))
    p = matrix(-np.ones((n, 1)))
    G = matrix(np.vstack((-np.eye(2*N), np.eye(2*N))))
    h = matrix(np.vstack((np.zeros((2*N, 1)), C*np.ones((2*N, 1)))))
    A = matrix(y_svm.reshape((1, -1)))
    b = matrix(np.zeros((1, 1)))
    solvers.options['show_progress'] = False
    sol = solvers.qp(Q, p, G, h, A, b)

    # List of lambdas and set of lambdas which are lambda > 0 or lambda < C
    l = np.array(sol['x']).reshape(2*N)
    S = np.where(l > 1e-5)[0]
    S2 = np.where(l < .999*C)[0]

    # Set of lambdas which are 0 < lambda < C 
    M = [val for val in S if val in S2]

    # Calculate W
    VS = V[S]
    lS = l[S]
    w_dual = lS.dot(VS)

    # Calculate b
    yM = y[M]
    XM = X[M]
    b_dual = np.mean(yM - XM.dot(w_dual))
    return w_dual, b_dual
```

```python
w, b = SVM_soft_margin(X, y, 100)
print(w.T)
print(b)

from sklearn.svm import SVC
model = SVC(kernel='linear', C=100)
model.fit(X, y)
w_sklearn = model.coef_.reshape(-1, 1)
b_sklearn = model.intercept_[0]
print(w_sklearn.T)
print(b_sklearn)
```

```python
plt.scatter(X0[:, 0], X0[:, 1], **square_setting)
plt.scatter(X1[:, 0], X1[:, 1], **circle_setting)
x_plot = np.linspace(1, 5, 50)
y_plot = (w[0] / -w[1]) * x_plot + (b / -w[1])

y_first_margin = (w[0] / -w[1]) * x_plot + ((b + 1) / -w[1])
y_second_margin = (w[0] / -w[1]) * x_plot + ((b - 1) / -w[1])

plt.plot(x_plot, y_plot, **main_line)
plt.plot(x_plot, y_first_margin, **supporting_line)
plt.plot(x_plot, y_second_margin, **supporting_line)
plt.fill_between(x_plot, 10, y_plot, facecolor='none', hatch='x', alpha=0.1)
plt.fill_between(x_plot, y_plot, facecolor='none', hatch='+', alpha=0.1)
plt.xlim(1, 5)
plt.ylim(0, 3)
```
