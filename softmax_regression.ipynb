```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
C, N = 5, 500 # number of classes and number of points per class
means = [[2, 2], [8, 3], [3, 6], [14, 2], [12, 8]]
cov = [[1, 0], [0, 1]]
x0 = np.random.multivariate_normal(means[0], cov, N)
x1 = np.random.multivariate_normal(means[1], cov, N)
x2 = np.random.multivariate_normal(means[2], cov, N)
x3 = np.random.multivariate_normal(means[3], cov, N)
x4 = np.random.multivariate_normal(means[4], cov, N)
X = np.concatenate([x0, x1, x2, x3, x4], axis=0)
X = np.concatenate([X, np.ones((X.shape[0], 1))], axis=1)
y = np.asarray([0]*N + [1]*N + [2]*N+ [3]*N + [4]*N)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Initial data")
```

```python
def softmax(W, X):
    res = X.dot(W)
    exp_res = np.exp(res - np.max(res, axis=1, keepdims=True))
    returned_res = exp_res / np.sum(exp_res, axis=1, keepdims=True)
    return returned_res

def softmax_grad(W, X, y):
    E = softmax(W, X)
    E[range(X.shape[0]), y] -= 1
    return (X.T).dot(E) / X.shape[0]

def softmax_loss(W, X, y):
    loss = softmax(W, X)
    return -np.mean(np.log(loss[range(X.shape[0]), y]))
```

```python
def softmax_fit(X, y, learning_rate=0.01, epochs=100, tolerance=1e-5, batch_size=10):
    N = X.shape[0]
    W = np.random.rand(X.shape[1], C)
    pre_W = W.copy()
    losses = [softmax_loss(W, X, y)]
    batch_num = int(np.ceil(N / batch_size))
    for it in range(epochs):
        ID_list = np.random.permutation(N)
        for num in range(batch_num):
            batch_id = ID_list[batch_size * num:min(batch_size * (num + 1), N)]
            X_batch = X[batch_id, :]
            y_batch = y[batch_id]
            W -= learning_rate * softmax_grad(W, X_batch, y_batch)
            losses.append(softmax_loss(W, X, y))
        if np.linalg.norm(W - pre_W) < tolerance:
            break
        pre_W = W.copy()

    return W, losses  
```

```python
res_W, loss_hist = softmax_fit(X, y)
```

```python
print(res_W)
```

```python
plt.plot(loss_hist, c='black', linewidth=1)
plt.title("Loss values")
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
x_test = np.array([[10, 8, 1]])
a_test = softmax(res_W, x_test)
print(np.argmax(a_test))
plt.scatter(x_test[0, 0], x_test[0, 1], c='red')
plt.text(2, 2, '0', c='red')
plt.text(8, 3, '1', c='red')
plt.text(3, 6, '2', c='red')
plt.text(14, 2, '3', c='red')
plt.text(12, 8, '4', c='red')
plt.title("Result in softmax regression")
```
